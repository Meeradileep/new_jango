```python
import numpy as np
marks = np.array([45,78,56,99,10])
print("Marks:",marks)
print("Lowest Marks:",np.min(marks))
print("Highest Marks:",np.max(marks))
print("Average Marks:",np.mean(marks))
print("Greated than 70:",marks[marks>70])
```
